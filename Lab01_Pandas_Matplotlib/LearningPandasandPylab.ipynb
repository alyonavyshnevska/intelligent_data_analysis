Add to the dataframe a new feature which will describe whether or not the user have done more then 3 service calls. Investigate how does this feature describes the churn rate.

```python
### Write Your code here ###
many_service_calls = []
for element in df['Customer service calls']:
    if element > 3:
        many_service_calls.append(1)
    else:
        many_service_calls.append(0)
        
df['Many Service Calls'] = many_service_calls
```

```python
df['new col'] = df['Customer service calls'].apply(lambda x: 0 if x < 4 else 1)
```
